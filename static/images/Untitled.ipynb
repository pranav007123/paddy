```python
from PIL import Image

def change_image_width(image_path, new_width, output_path):
    """
    Change the width of an image while maintaining aspect ratio.
    
    Args:
    image_path (str): Path to the input image.
    new_width (int): New width for the image.
    output_path (str): Path to save the modified image.
    """
    # Open the image
    with Image.open(image_path) as img:
        # Calculate the aspect ratio
        aspect_ratio = img.width / img.height
        # Calculate the new height
        new_height = int(new_width / aspect_ratio)
        
        # Resize the image
        resized_img = img.resize((new_width, new_height), Image.ANTIALIAS)
        
        # Save the resized image
        resized_img.save(output_path)

# Example usage:
input_image_path = "DD.jpeg"
output_image_path = "output_image.jpg"
new_width = 480  # Set the new width here
change_image_width(input_image_path, new_width, output_image_path)
```
